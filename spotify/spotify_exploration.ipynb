### Let's crete a pie chart of the top 10 popular artists. The chart shows how many of their songs are there in this dataset. It is not a measure of their popularity, rather how prolific they were (at least according to this dataset).

```python
top_10_artists_df = df_artist.sort_values(by=['count'], ascending=False).head(10)
df_plt = pd.DataFrame({'y': top_10_artists_df['count'].values},index = top_10_artists_df['artists'].values)

df_plt.plot.pie(y='y',figsize=(12,12))
```
